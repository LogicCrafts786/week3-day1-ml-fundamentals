## ðŸ“¦ Section 2: Import Libraries

We start by importing all the libraries needed for this project:
- **Pandas**: Data loading and manipulation
- **NumPy**: Numerical computations
- **Plotly Express**: Interactive visualizations
- **Scikit-learn**: ML model, splitting, and evaluation metrics

```python
# â”€â”€ Import all required libraries â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import pandas as pd                          # For data manipulation
import numpy as np                           # For numerical operations
import plotly.express as px                  # For interactive visualizations

from sklearn.model_selection import train_test_split   # To split data
from sklearn.linear_model import LinearRegression      # Our ML model
from sklearn.metrics import mean_absolute_error, r2_score  # Evaluation metrics

print("âœ… All libraries imported successfully!")
```

## ðŸ“‚ Section 3: Load Dataset

We load our CSV file using Pandas and explore its basic structure.

```python
# â”€â”€ Load the dataset from CSV â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
df = pd.read_csv('student_performance.csv')

print("âœ… Dataset loaded successfully!\n")

# Display first 5 rows to preview the data
print("ðŸ“‹ First 5 Rows:")
display(df.head())
```

```python
# â”€â”€ Dataset Shape (rows x columns) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print(f"ðŸ“ Dataset Shape: {df.shape}")
print(f"   â†’ {df.shape[0]} rows (students) and {df.shape[1]} columns (features)\n")

# â”€â”€ Data types of each column â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("ðŸ“Š Data Types:")
print(df.dtypes)
print()
```

```python
# â”€â”€ Basic statistical summary â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("ðŸ“ˆ Basic Statistics:")
display(df.describe())
```

## ðŸ” Section 4: Dataset Exploration

Let's explore the dataset deeper to understand its structure.

```python
# â”€â”€ Comprehensive dataset exploration â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("=" * 50)
print("       DATASET EXPLORATION SUMMARY")
print("=" * 50)

print(f"\nðŸ“Œ Number of Rows    : {df.shape[0]}")
print(f"ðŸ“Œ Number of Columns : {df.shape[1]}")

# Check for missing values
print(f"\nâ“ Missing Values per Column:")
print(df.isnull().sum())

# Determine if dataset is clean
total_missing = df.isnull().sum().sum()
if total_missing == 0:
    print("\nâœ… Great! No missing values found. Dataset is clean.")
else:
    print(f"\nâš ï¸ Warning: {total_missing} missing values found. Handle before training.")

print("\nðŸ“‹ Column Summary:")
for col in df.columns:
    print(f"  â€¢ {col}: min={df[col].min()}, max={df[col].max()}, mean={df[col].mean():.2f}")
```

## ðŸŽ¯ Section 5: Features and Labels

In supervised learning:
- **Features (X):** The input variables the model uses to learn. We chose:
  - `Hours_Studied` â€” More study time generally leads to better scores
  - `Attendance_Percentage` â€” Regular attendance helps students understand material
  - `Previous_Score` â€” Past performance is a strong predictor of future performance
- **Label (y):** The output variable we want to predict â†’ `Score`

These features are chosen because they have a logical, real-world relationship with the final score.

```python
# â”€â”€ Define Features (X) and Label (y) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

# Features: input variables used for prediction
X = df[['Hours_Studied', 'Attendance_Percentage', 'Previous_Score']]

# Label: the output we want to predict
y = df['Score']

print("âœ… Features (X) defined:")
print(f"   Columns: {list(X.columns)}")
print(f"   Shape  : {X.shape}\n")

print("âœ… Label (y) defined:")
print(f"   Column : Score")
print(f"   Shape  : {y.shape}")
```

## ðŸ”„ Section 6: Machine Learning Workflow

Here is the complete ML workflow we will follow in this project:

| Step | Description |
|------|-------------|
| 1ï¸âƒ£ Data Collection | Gathered student data with 4 columns and 110 records |
| 2ï¸âƒ£ Data Cleaning | Verified no missing values; data is ready to use |
| 3ï¸âƒ£ Feature Selection | Chose 3 relevant input features; 1 target label |
| 4ï¸âƒ£ Train-Test Split | 80% for training, 20% for testing (random_state=42) |
| 5ï¸âƒ£ Model Training | Fit a Linear Regression model on training data |
| 6ï¸âƒ£ Prediction | Use trained model to predict test set scores |
| 7ï¸âƒ£ Evaluation | Measure accuracy using MAE and RÂ² Score |
| 8ï¸âƒ£ Visualization | Plot relationships and distributions using Plotly |

---
Linear Regression learns this formula:

> **Score = (c1 Ã— Hours_Studied) + (c2 Ã— Attendance_Percentage) + (c3 Ã— Previous_Score) + intercept**

## âœ‚ï¸ Section 7: Train-Test Split

We split our data so the model trains on 80% and is evaluated on the unseen 20%.

```python
# â”€â”€ Split data into Training (80%) and Testing (20%) sets â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% data goes to testing
    random_state=42      # Ensures reproducible results every time
)

print("âœ… Train-Test Split Complete!")
print("=" * 40)
print(f"ðŸ“š Total   samples : {len(X)}")
print(f"ðŸ‹ï¸ Training samples: {len(X_train)} ({len(X_train)/len(X)*100:.0f}%)")
print(f"ðŸ§ª Testing  samples: {len(X_test)} ({len(X_test)/len(X)*100:.0f}%)")
```

## ðŸ‹ï¸ Section 8: Train Linear Regression Model

We create a Linear Regression model and fit it on the training data.

```python
# â”€â”€ Create and train the Linear Regression model â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
model = LinearRegression()   # Initialize the model
model.fit(X_train, y_train)  # Train on the training data

print("âœ… Model trained successfully!\n")

# â”€â”€ Display the learned coefficients and intercept â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
print("ðŸ“Š Model Coefficients (how much each feature affects Score):")
print("-" * 55)
for feature, coef in zip(X.columns, model.coef_):
    print(f"  {feature:<28}: {coef:.4f}")

print(f"\n  Intercept (base score)        : {model.intercept_:.4f}")
print("\nðŸ’¡ Interpretation:")
print("  Each coefficient tells us: for a 1-unit increase in that")
print("  feature, the predicted Score changes by that coefficient amount,")
print("  while keeping other features constant.")
```

## ðŸ”® Section 9: Predictions on Test Data

Now we use the trained model to predict scores for the test set.

```python
# â”€â”€ Predict on the test set â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
y_pred = model.predict(X_test)

print("âœ… Predictions generated for test set!\n")
print("ðŸ”¢ First 10 Predicted Scores:")
print("-" * 30)
for i, pred in enumerate(y_pred[:10], 1):
    print(f"  Student {i:2d}: Predicted Score = {pred:.2f}")
```

## ðŸ†• Section 10: Predict Scores for New Students

Let's predict scores for 3 hypothetical new students using our trained model.

```python
# â”€â”€ Define new students' data â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
new_students = pd.DataFrame({
    'Hours_Studied'         : [8, 5, 10],
    'Attendance_Percentage' : [90, 75, 95],
    'Previous_Score'        : [80, 65, 88]
})

student_names = ['Student A', 'Student B', 'Student C']

# â”€â”€ Predict their scores â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
new_predictions = model.predict(new_students)

print("ðŸŽ“ Predictions for New Students")
print("=" * 60)
print(f"{'Name':<12} {'Hours':>6} {'Attend%':>8} {'Prev Score':>11} {'Predicted Score':>16}")
print("-" * 60)
for name, (_, row), pred in zip(student_names, new_students.iterrows(), new_predictions):
    print(f"{name:<12} {row['Hours_Studied']:>6} {row['Attendance_Percentage']:>8} {row['Previous_Score']:>11} {pred:>16.2f}")

print("\nðŸ’¡ Higher study hours, attendance, and previous score â†’ higher predicted score")
```

## ðŸ“‹ Section 11: Actual vs Predicted Comparison

We compare the model's predictions against the actual test values.

```python
# â”€â”€ Build comparison DataFrame â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
comparison_df = pd.DataFrame({
    'Actual Score'   : y_test.values,
    'Predicted Score': np.round(y_pred, 2),
    'Error'          : np.round(np.abs(y_test.values - y_pred), 2)
})

comparison_df.index = range(1, len(comparison_df) + 1)

print("ðŸ“Š Actual vs Predicted Scores (First 20 rows):")
display(comparison_df.head(20))
```

## ðŸ“ Section 12: Model Evaluation

We evaluate our model using two key metrics:
- **MAE (Mean Absolute Error):** Average prediction error in score points
- **RÂ² Score:** Proportion of variance explained by the model (0 to 1)

```python
# â”€â”€ Calculate evaluation metrics â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
mae = mean_absolute_error(y_test, y_pred)
r2  = r2_score(y_test, y_pred)

print("=" * 50)
print("       MODEL EVALUATION RESULTS")
print("=" * 50)
print(f"\nðŸ“‰ Mean Absolute Error (MAE) : {mae:.4f} points")
print(f"ðŸ“ˆ RÂ² Score                  : {r2:.4f}")

print("\nðŸ’¡ What these mean:")
print(f"  â€¢ MAE = {mae:.2f} means our predictions are off by only ~{mae:.1f} points on average.")
print(f"  â€¢ RÂ² = {r2:.4f} means the model explains {r2*100:.1f}% of the variance in scores.")

# Performance interpretation
if r2 >= 0.95:
    print("\n  ðŸ† Excellent model performance!")
elif r2 >= 0.85:
    print("\n  âœ… Good model performance.")
else:
    print("\n  âš ï¸ Model could be improved with more features or data.")
```

## ðŸ“Š Section 13: Data Visualization Using Plotly

We create 5 interactive visualizations to understand the data and model performance.

```python
# â”€â”€ Plot 1: Scatter Plot â€” Hours Studied vs Score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig1 = px.scatter(
    df,
    x='Hours_Studied',
    y='Score',
    title='ðŸ“š Hours Studied vs Score',
    labels={'Hours_Studied': 'Hours Studied', 'Score': 'Exam Score'},
    color='Score',
    color_continuous_scale='Blues',
    trendline='ols',        # Adds regression trendline
    template='plotly_white'
)
fig1.update_traces(marker=dict(size=9, opacity=0.8))
fig1.update_layout(
    title_font_size=18,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14
)
fig1.show()
print("ðŸ’¡ Observation: Students who study more hours tend to score higher.")
```

```python
# â”€â”€ Plot 2: Scatter Plot â€” Attendance vs Score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig2 = px.scatter(
    df,
    x='Attendance_Percentage',
    y='Score',
    title='ðŸ“… Attendance Percentage vs Score',
    labels={'Attendance_Percentage': 'Attendance (%)', 'Score': 'Exam Score'},
    color='Score',
    color_continuous_scale='Greens',
    trendline='ols',
    template='plotly_white'
)
fig2.update_traces(marker=dict(size=9, opacity=0.8))
fig2.update_layout(title_font_size=18)
fig2.show()
print("ðŸ’¡ Observation: Higher attendance is associated with higher exam scores.")
```

```python
# â”€â”€ Plot 3: Scatter Plot â€” Previous Score vs Score â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig3 = px.scatter(
    df,
    x='Previous_Score',
    y='Score',
    title='ðŸ“ Previous Score vs Current Score',
    labels={'Previous_Score': 'Previous Exam Score', 'Score': 'Current Exam Score'},
    color='Score',
    color_continuous_scale='Reds',
    trendline='ols',
    template='plotly_white'
)
fig3.update_traces(marker=dict(size=9, opacity=0.8))
fig3.update_layout(title_font_size=18)
fig3.show()
print("ðŸ’¡ Observation: Previous score is the strongest predictor of current score.")
```

```python
# â”€â”€ Plot 4: Histogram â€” Score Distribution â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
fig4 = px.histogram(
    df,
    x='Score',
    nbins=20,
    title='ðŸ“Š Distribution of Exam Scores',
    labels={'Score': 'Exam Score', 'count': 'Number of Students'},
    color_discrete_sequence=['steelblue'],
    template='plotly_white'
)
fig4.update_layout(
    title_font_size=18,
    xaxis_title='Exam Score',
    yaxis_title='Number of Students',
    bargap=0.1
)
fig4.show()
print("ðŸ’¡ Observation: Scores are spread across a range, showing varied student performance.")
```

```python
# â”€â”€ Plot 5: Correlation Heatmap â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import plotly.graph_objects as go

# Calculate correlation matrix
corr_matrix = df.corr()

fig5 = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns.tolist(),
    y=corr_matrix.columns.tolist(),
    colorscale='RdBu',
    zmin=-1, zmax=1,
    text=np.round(corr_matrix.values, 2),
    texttemplate='%{text}',
    textfont={"size": 14},
    hoverongaps=False
))

fig5.update_layout(
    title='ðŸ”¥ Feature Correlation Heatmap',
    title_font_size=18,
    template='plotly_white',
    width=600,
    height=500
)
fig5.show()
print("ðŸ’¡ Observation: All features show strong positive correlation with Score.")
print("   Values close to 1 = strong positive correlation.")
```
